## **Cargar datos.**

```python
# cargar datos "train_EDA", "test_EDA", y "muestra"
train = pd.read_csv('../data/train_EDA_1.csv')
test = pd.read_csv('../data/test_EDA_1.csv')
muestra = pd.read_csv('../data/muestra.csv')
```

## **Training.**

```python
# dividir muestras
X = train.drop('Price', axis=1)
y = train.Price

X.shape, y.shape
```

```python
# dividir datos
X_train, X_test, y_train, y_test = tts(X, y, train_size=0.8, test_size=0.2, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# iniciar "LazyPredict"
lazy=LazyRegressor()

lazy.fit(X_train, X_test, y_train, y_test)
```

```python
# aplicar modelo segun resultados de "LazyPredict"
gbr.fit(X, y)
```

```python
# realizar prediccion
y_pred_1 = gbr.predict(test)

y_pred_1
```

```python
# exportar resultados a "muestra"
muestra['Price']= y_pred_1
```
